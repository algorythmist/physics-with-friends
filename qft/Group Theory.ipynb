# Group Representations

### Basic group-theoretic definitions

A **group homomorphism** is a function
$$
f : G_1 \to G_2
$$
between groups that preserves the group operation:
$$
f(gh) = f(g)\,f(h)
\quad \text{for all } g,h \in G_1.
$$

A **group isomorphism** is a **bijective** group homomorphism.  
Two groups are said to be *isomorphic* if there exists an isomorphism between them.

---

An **automorphism** of a set $S$ is a bijection
$$
f : S \to S.
$$
The set of all automorphisms of $S$ forms a group under composition, with inverses given by inverse functions.  
This group is denoted by
$$
\mathrm{Aut}(S).
$$

---

A **group action** of a group $G$ on a set $S$ is a group homomorphism
$$
\varphi : G \to \mathrm{Aut}(S).
$$


### Linear representations of groups

Let $V$ be a vector space over a field $\mathbb{F}$.  
The set of all invertible linear transformations of $V$ is denoted by
$$
GL(V,\mathbb{F}).
$$
This set forms a group under composition, with identity given by the identity map on $V$ and inverses given by inverse linear transformations.

---

A **representation** of a group $G$ on a vector space $V$ is a group homomorphism
$$
\rho : G \to GL(V,\mathbb{F}).
$$

---

A subspace $W \subset V$ is said to be **invariant under $\rho$** if
$$
\rho(g)(W) \subset W \quad \text{for all } g \in G.
$$

If $W$ is invariant, then the restriction of $\rho$ to $W$ defines a representation of $G$ on $W$.

---

A representation $\rho$ is called **reducible** if it has a nontrivial invariant subspace, i.e., a subspace
$$
W \subset V \quad \text{with} \quad \{0\} \neq W \neq V.
$$

A representation $\rho$ is called **irreducible** if its only invariant subspaces are $\{0\}$ and $V$ itself.


The representation $\rho$ is called **faithful** if it is **injective**.

---



## Induced representation on a function space

Let $V$ be a finite-dimensional vector space over $\mathbb{R}$ (or $\mathbb{C}$), and let
$$
L : G \to \mathrm{GL}(V)
$$
be a linear representation of a group $G$ on $V$.

Let $H = L^2(V)$ be the Hilbert space of square-integrable functions
$$
f : V \to \mathbb{C},
$$
with inner product defined using the Lebesgue measure $dx$ on $V$:
$$
\langle f, h \rangle
= \int_V f(x)\,\overline{h(x)}\,dx.
$$

---

### Induced representation

The representation $L$ induces a representation
$$
\widetilde{L} : G \to \mathcal{B}(H)
$$
on $H$ defined by pullback:
$$
(\widetilde{L}(g)f)(x) := f\big(L(g^{-1})x\big).
$$

---

### Representation property

For $g,h \in G$ and $f \in H$,
$$
\begin{aligned}
(\widetilde{L}(gh)f)(x)
&= f\big(L((gh)^{-1})x\big) \\
&= f\big(L(h^{-1})L(g^{-1})x\big) \\
&= (\widetilde{L}(g)\widetilde{L}(h)f)(x).
\end{aligned}
$$

Thus,
$$
\widetilde{L}(gh) = \widetilde{L}(g)\widetilde{L}(h),
$$
so $\widetilde{L}$ is a representation of $G$ on $H$.

---

### Behavior of the norm

The $L^2$ norm of $\widetilde{L}(g)f$ is
$$
\|\widetilde{L}(g)f\|^2
= \int_V |f(L(g^{-1})x)|^2\,dx.
$$

Using the change of variables $y = L(g^{-1})x$, we have
$$
dx = |\det L(g)|\,dy,
$$
and therefore
$$
\|\widetilde{L}(g)f\|^2
= |\det L(g)| \int_V |f(y)|^2\,dy
= |\det L(g)|\,\|f\|^2.
$$

---

### Unitarity

- If $|\det L(g)| = 1$ for all $g \in G$ (for example, if $L(g)$ is orthogonal or unitary),
  then $\widetilde{L}$ is a unitary representation.

- In general, one can define a normalized action
  $$
  (\widetilde{L}(g)f)(x)
  := |\det L(g)|^{-1/2} f\big(L(g^{-1})x\big),
  $$
  which satisfies
  $$
  \|\widetilde{L}(g)f\| = \|f\|.
  $$

This normalization yields a unitary representation of $G$ on $H$.

---

### Interpretation

This construction is the natural action of $G$ on functions induced by its linear action on $V$:
- $G$ acts on points of $V$ via $L$
- functions transform by pullback
- the determinant accounts for the change of volume

---

### Summary

A linear representation $L : G \to \mathrm{GL}(V)$ induces a representation on $L^2(V)$ by
$$
(\widetilde{L}(g)f)(x) = f(L(g^{-1})x),
$$
which is unitary exactly when the action preserves Lebesgue measure (or after determinant normalization).


## Induced action on operators on $H$

Let $H = L^2(V)$ and let
$$
\widetilde{L} : G \to \mathcal{U}(H)
$$
be the (unitary) representation induced by the linear action
$L : G \to \mathrm{GL}(V)$, given by
$$
(\widetilde{L}(g)f)(x)
= |\det L(g)|^{-1/2} f\big(L(g^{-1})x\big).
$$

Let $\mathcal{B}(H)$ denote the space of bounded linear operators on $H$
(or any $G$-stable subspace such as trace-class or Hilbert–Schmidt operators).

---

### Induced representation on operators

The representation $\widetilde{L}$ induces a representation
$$
\mathrm{Ad}_{\widetilde{L}} : G \to \mathrm{Aut}(\mathcal{B}(H))
$$
defined by **conjugation**:
$$
(\mathrm{Ad}_{\widetilde{L}}(g)A)
:= \widetilde{L}(g)\, A \,\widetilde{L}(g)^{-1},
\qquad A \in \mathcal{B}(H).
$$

### Why operators transform by conjugation

The action of $G$ on operators must be compatible with its action on states.
If $f \in H$ is a state and $A \in \mathcal{B}(H)$ is an operator, then applying
$A$ and then transforming by $g \in G$ should give the same result as first
transforming the state and then applying the transformed operator.

This covariance requirement is expressed by
$$
(\mathrm{Ad}_{\widetilde{L}}(g)A)\big(\widetilde{L}(g)f\big)
= \widetilde{L}(g)(Af)
\quad \text{for all } f \in H.
$$

Rewriting this condition gives
$$
\widetilde{L}(g)^{-1}(\mathrm{Ad}_{\widetilde{L}}(g)A)\widetilde{L}(g) = A,
$$
which uniquely determines the transformation rule
$$
\mathrm{Ad}_{\widetilde{L}}(g)A
= \widetilde{L}(g)\,A\,\widetilde{L}(g)^{-1}.
$$

---

### Representation property

For $g,h \in G$ and $A \in \mathcal{B}(H)$,
$$
\begin{aligned}
\mathrm{Ad}_{\widetilde{L}}(gh)(A)
&= \widetilde{L}(gh) A \widetilde{L}(gh)^{-1} \\
&= \widetilde{L}(g)\big(\widetilde{L}(h) A \widetilde{L}(h)^{-1}\big)\widetilde{L}(g)^{-1} \\
&= \mathrm{Ad}_{\widetilde{L}}(g)\big(\mathrm{Ad}_{\widetilde{L}}(h)(A)\big).
\end{aligned}
$$

Thus,
$$
\mathrm{Ad}_{\widetilde{L}}(gh)
= \mathrm{Ad}_{\widetilde{L}}(g)\,\mathrm{Ad}_{\widetilde{L}}(h),
$$
so $\mathrm{Ad}_{\widetilde{L}}$ is a representation of $G$ on $\mathcal{B}(H)$.

---


### Invariant operators

An operator $A \in \mathcal{B}(H)$ is $G$-invariant if
$$
\widetilde{L}(g) A = A \widetilde{L}(g)
\quad \text{for all } g \in G.
$$

Equivalently,
$$
\mathrm{Ad}_{\widetilde{L}}(g)(A) = A.
$$

Such operators form the **commutant** of the representation $\widetilde{L}$.

---

### Interpretation

- The induced action on $H$ describes how functions transform.
- The induced action on $\mathcal{B}(H)$ describes how **observables, kernels,
  and operators** transform.
- This is the natural “change of coordinates” action on operators.

---

### Summary

Given a representation $\widetilde{L}$ of $G$ on $H$, the induced representation
on operators is
$$
A \mapsto \widetilde{L}(g)\,A\,\widetilde{L}(g)^{-1}.
$$

This conjugation action governs the transformation of kernels, observables,
and invariant operators under the symmetry group $G$.


# Lie Groups

### Topological groups and Lie groups

A **topological group** is a group $G$ equipped with a topology $\tau$ such that
the group operations
$$
G \times G \to G, \quad (g,h) \mapsto gh,
$$
and
$$
G \to G, \quad g \mapsto g^{-1},
$$
are continuous with respect to $\tau$.

---

A **Lie group** is a topological group $G$ that is also a smooth (i.e. $C^\infty$)
manifold of finite dimension $n$, such that the group operations are smooth maps.

Lie groups admit **finite-dimensional linear representations**, meaning there
exist homomorphisms
$$
\rho : G \to GL(V)
$$
for some finite-dimensional vector space $V$.
In particular, every Lie group has a faithful local realization by matrices
acting on its Lie algebra via the adjoint representation.

---



### Generators, exponentials, and the Lie algebra


A subset $K \subset G$ is said to **generate** a group $G$ if every element
$g \in G$ can be written as a finite product of elements of $K$ and their inverses.

---

**Fact 1.**  
If $G$ is a connected Lie group, then **any neighborhood of the identity**
$e \in G$ generates $G$.

---

**Fact 2.**  
Every element of a connected Lie group $G$ can be written (at least locally, and
globally for connected components) in the form
$$
g = \exp\!\left(\theta_i X^i\right),
$$
where $\{X^i\}$ is a basis of the tangent space $T_e G$ and $\theta_i \in \mathbb{R}$.

### From group structure to Lie algebra structure

The tangent space at the identity,
$$
\mathfrak{g} := T_e G,
$$
captures the **infinitesimal structure** of the Lie group.

Because any neighborhood of the identity generates $G$, understanding how group
elements behave *arbitrarily close to the identity* is sufficient to recover the
entire group structure. Thus, the local behavior near $e$ encodes the global
structure of $G$.

The exponential map
$$
\exp : T_e G \to G
$$
associates to each tangent vector $X \in T_e G$ a one-parameter subgroup
$$
t \mapsto \exp(tX),
$$
representing an infinitesimal motion integrated into a finite group element.

Since group multiplication is generally noncommutative, the product
$$
\exp(tX)\exp(sY)
$$
does not equal $\exp(sY)\exp(tX)$ in general. The leading-order difference between
these products, when expanded for small $t$ and $s$, defines a bilinear operation
$$
[X,Y] \in T_e G.
$$

This operation measures the infinitesimal failure of group elements to commute and
is inherited directly from the group structure. It satisfies:
- bilinearity,
- antisymmetry,
- the Jacobi identity,

and therefore defines a **Lie bracket** on $T_e G$.

Hence, the tangent space at the identity naturally acquires the structure of a
Lie algebra, called the **Lie algebra of $G$**, denoted
$$
\mathfrak{g} = T_e G.
$$


### Infinitesimal elements and the Lie bracket

Let $G$ be a Lie group and $\mathfrak{g} = T_e G$ its tangent space at the identity.

---

#### 1. Infinitesimal elements

Consider an element **infinitesimally close to the identity**:
$$
g = \exp(i \epsilon X) \approx I + i \epsilon X,
$$
where $X \in \mathfrak{g}$ and $\epsilon \ll 1$ is a small real parameter.

Similarly, for another element $h \in G$:
$$
h = \exp(i \epsilon Y) \approx I + i \epsilon Y,
$$
with $Y \in \mathfrak{g}$.

---

#### 2. Commutator of infinitesimal elements

The product $ghg^{-1}h^{-1}$ measures the **noncommutativity** of $g$ and $h$:

\begin{aligned}
ghg^{-1}h^{-1} &\approx (I + i \epsilon X)(I + i \epsilon Y)(I - i \epsilon X)(I - i \epsilon Y) \\
&= I + (i \epsilon)^2 (XY - YX) + O(\epsilon^3) \\
&= I - \epsilon^2 [X,Y] + O(\epsilon^3),
\end{aligned}


where we define the **Lie bracket**
$$
[X,Y] := XY - YX.
$$

---

#### 3. Why $[X,Y] \in \mathfrak{g}$

- The tangent space $\mathfrak{g}$ is **the space of infinitesimal generators**.
- The commutator $[X,Y]$ arises as the **leading-order term** of the group commutator near the identity.
- Since the commutator of elements of $G$ near the identity is itself infinitesimal, $[X,Y]$ must be another tangent vector at the identity:
$$
[X,Y] \in T_e G = \mathfrak{g}.
$$

Thus the bracket **closes in the tangent space**, giving a well-defined Lie algebra structure.

---

#### 4. Summary

- Infinitesimal elements: $g \approx I + i \epsilon X$.
- Group multiplication induces a commutator:
$$
ghg^{-1}h^{-1} \approx I - \epsilon^2 [X,Y].
$$
- The **Lie bracket** $[X,Y]$ measures the infinitesimal noncommutativity of $G$.
- Closure of $[X,Y]$ in $\mathfrak{g}$ is guaranteed because the commutator of infinitesimal group elements is itself infinitesimal.

This is the **tangent-space manifestation of the group structure**, which is exactly what makes $T_e G$ into a Lie algebra.


### Structure constants of a Lie algebra

Let $\mathfrak{g} = T_e G$ be the Lie algebra of a Lie group $G$, and let
$$
\{X^i\}_{i=1}^n
$$
be a basis of $\mathfrak{g}$ (dimension $n$).

---

#### 1. Expressing brackets in a basis

Since the Lie bracket is bilinear and closed in $\mathfrak{g}$, the bracket of two basis elements
can be written as a linear combination of the basis:
$$
[X^i, X^j] = f^{ij}{}_k \, X^k,
$$
where the coefficients $f^{ij}{}_k \in \mathbb{R}$ (or $\mathbb{C}$) are called the **structure constants**.

---

#### 2. How structure constants arise

1. Consider two infinitesimal elements in the group:
$$
g = I + i \epsilon X^i, \qquad
h = I + i \epsilon X^j, \quad \epsilon \ll 1.
$$
2. The group commutator to leading order gives:
$$
ghg^{-1}h^{-1} \approx I - \epsilon^2 [X^i, X^j].
$$
3. Since $[X^i, X^j] \in \mathfrak{g}$, it can be expanded in the basis:
$$
[X^i, X^j] = f^{ij}{}_k X^k.
$$
Thus the **structure constants** $f^{ij}{}_k$ encode the infinitesimal noncommutativity of the group.

---

#### 3. Properties of structure constants

From the properties of the Lie bracket:

1. **Antisymmetry**:
$$
f^{ij}{}_k = -f^{ji}{}_k
$$
because $[X^i,X^j] = -[X^j,X^i]$.

2. **Jacobi identity**:
$$
[X^i,[X^j,X^k]] + [X^j,[X^k,X^i]] + [X^k,[X^i,X^j]] = 0
$$
implies a quadratic relation on $f^{ij}{}_k$:
$$
f^{im}{}_l f^{jk}{}_m + f^{jm}{}_l f^{ki}{}_m + f^{km}{}_l f^{ij}{}_m = 0.
$$

These constraints reflect the underlying **group structure** in the algebra.

---

#### 4. Intuition

- The **structure constants** tell us how infinitesimal motions along basis directions fail to commute.
- They are completely determined by the **local group multiplication near the identity**.
- In physics, they appear as the **commutation relations of generators** of symmetries.
