### Problem 22.2 

Use the canonical commutation relations to verify equation 22.38.

#### Solution

$$
P^\mu := \int d^3x T^{0\mu}(x)
$$

where 

$$
T^{0j} = -\Pi\partial^j \phi
$$

and 

$$
T^{00} = \frac{1}{2}\Pi^2 + \frac{1}{2}(\nabla\phi)^2+V(\phi)
$$

Therefore

\begin{equation}
[\phi(x), P^j] = -\int d^3x' \, [\phi(x), \Pi(x') \partial^j \phi(x')] = -\int d^3x' \, [\phi(x), \Pi(x')] \nabla^j \phi(x')
\end{equation}

since $\partial^j \phi(x')$ commutes with $\phi(x)$. 

Use the canonical quantization relations:

\begin{equation}
[\phi(x), P^j] = -i \int d^3x' \, \delta^3(x - x') \nabla^j \phi(x') = -i \partial^j \phi(x)
\end{equation}



\begin{equation}
[\phi(x), P^0] = [\phi(x), \int d^3x'T^{00}(x')] = \int d^3x' \, \left[\phi(x), \frac{1}{2}\Pi^2(x') + \frac{1}{2}(\nabla\phi)^2 + V(\phi)\right]
\end{equation}

$\phi$ commutes with $\nabla\phi$ and $V(\phi)$, so

\begin{equation}
[\phi(x), P^0] = \frac{1}{2} \int d^3x' \, [\phi(x), \Pi^2(x')]
\end{equation}

which gives, using the properties of commutators:
\begin{equation}
[\phi(x), P^0] = \frac{1}{2} \int d^3x' \, [\phi(x), \Pi(x')] \Pi(x') + \frac{1}{2} \int d^3x' \, \Pi(x') [\phi(x), \Pi(x')]  = \int d^3x' \, i\delta^3(x - x') \Pi(x') = i\Pi(x) = i\dot{\phi}(x)
\end{equation}

Putting it all togethe

\begin{equation}
\boxed{[\phi(x), P^\mu] = -i \partial^\mu \phi(x)}
\end{equation}

as expected.
