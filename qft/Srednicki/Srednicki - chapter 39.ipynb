### Exercise 39.1

Verify 39.24



#### Solution

plugining in the general solution for $\Psi$:

\begin{align}
Q &= \int d^3x\, \overline{\Psi}\gamma^0 \Psi \\
  &= \int d^3x\, \widetilde{dp}\, \widetilde{dp'}\,
     \Big[ b_{s'}^\dagger(\mathbf{p'})\, \overline{u}_{s'}(\mathbf{p'})\, e^{-ip'x}
         + d_{s'}(\mathbf{p'})\, \overline{v}_{s'}(\mathbf{p'})\, e^{+ip'x} \Big] \gamma^0 \nonumber \\
  &\qquad \times
     \Big[ b_s(\mathbf{p})\, u_s(\mathbf{p})\, e^{+ipx}
         + d_s^\dagger(\mathbf{p})\, v_s(\mathbf{p})\, e^{-ipx} \Big] \,.
\end{align}

Let's look at the first term:

$$
\int \widetilde{dp'}\, d^3x\;
  b_{s'}^\dagger(\mathbf{p'})\, \overline{u}_{s'}(\mathbf{p'})\, \gamma^0\, b_s(\mathbf{p})\, u_s(\mathbf{p})\,
  e^{i(p - p')x}
= (2\pi)^3 \int \widetilde{dp'}\;
  b_{s'}^\dagger(\mathbf{p'})\, \overline{u}_{s'}(\mathbf{p'})\, \gamma^0\, b_s(\mathbf{p})\, u_s(\mathbf{p})\,
  \delta^3(\mathbf{p} - \mathbf{p'}) \,.
$$

where I have used

$$
\int d^3 x\; e^{ip\cdot x} = (2\pi)^3\delta^3(p)
$$


Now remembering 

$$
\tilde{dp} = \frac{d^3 p}{2\omega(2\pi)^3}
$$

and 

\begin{align}
\overline{u}_{s'}(\mathbf{p}) \gamma^\mu u_s(\mathbf{p}) &= 2p^\mu \delta_{s's} \,, \\
\overline{v}_{s'}(\mathbf{p}) \gamma^\mu v_s(\mathbf{p}) &= 2p^\mu \delta_{s's} \,.
\end{align}


this is equal to

$$
\frac{1}{2\omega}\, b_{s'}^\dagger(\mathbf{p})\, b_s(\mathbf{p})\,
  \overline{u}_{s'}(\mathbf{p})\, \gamma^0\, u_s(\mathbf{p})
= \frac{1}{2\omega}\, b_{s'}^\dagger(\mathbf{p})\, b_s(\mathbf{p})\, 2\omega\, \delta_{s's}
= b_s^\dagger(\mathbf{p})\, b_s(\mathbf{p}) \,.
$$

Similarily, the last term gives

$$
\int \widetilde{dp'}\, d^3x\;
  d_{s'}(\mathbf{p'})\, \overline{v}_{s'}(\mathbf{p'})\, \gamma^0\, d_s^\dagger(\mathbf{p})\, v_s(\mathbf{p})\,
  e^{i(p' - p)x}
= d_s(\mathbf{p})\, d_s^\dagger(\mathbf{p}) \,.
$$

The second term is

\begin{align}
&\int \widetilde{dp'}\, d^3x\;
  b_{s'}^\dagger(\mathbf{p'})\, \overline{u}_{s'}(\mathbf{p'})\, \gamma^0\, d_s^\dagger(\mathbf{p})\, v_s(\mathbf{p})\,
  e^{i(p + p')x} \nonumber \\
&\qquad = (2\pi)^3 \int \widetilde{dp'}\;
  b_{s'}^\dagger(\mathbf{p'})\, \overline{u}_{s'}(\mathbf{p'})\, \gamma^0\, d_s^\dagger(\mathbf{p})\, v_s(\mathbf{p})\,
  \delta^3(\mathbf{p} + \mathbf{p'})  e^{2i\omega t}\\
&\qquad = \frac{1}{2\omega}\, b_{s'}^\dagger(-\mathbf{p})\, d_s^\dagger(\mathbf{p})\,
  \overline{u}_{s'}(-\mathbf{p})\, \gamma^0\, v_s(\mathbf{p}) \, e^{2i\omega t}.
\end{align}

and remembering that 

\begin{align}
\overline{u}_{s'}(\mathbf{p})\, \gamma^0\, v_s(-\mathbf{p}) &= 0 \,, \\
\overline{v}_{s'}(\mathbf{p})\, \gamma^0\, u_s(-\mathbf{p}) &= 0 \,.
\end{align}

We conclude that the entire term is 0.

and similarly, the third term is 0

So, what remains is

$$
Q = \sum_s \int \widetilde{dp'}\, \left[ b_s^\dagger(\mathbf{p})\, b_s(\mathbf{p}) \,. + d_s(\mathbf{p})\, d_s^\dagger(\mathbf{p}) \,. \right]
$$

After normal ordering

$$
Q = \sum_s \int \widetilde{dp'}\, \left[ b_s^\dagger(\mathbf{p})\, b_s(\mathbf{p}) \,. + d^\dagger_s(\mathbf{p})\, d_s(\mathbf{p}) \,. \right]
$$
