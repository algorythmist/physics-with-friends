## Notation and Properties

$$
\beta \equiv \begin{bmatrix}0 & I\\I & 0\end{bmatrix}
$$

$$
\beta^T = \beta^\dagger = \beta^{-1} = \beta
$$

For an operator $A$, we define

$$
\overline{A} \equiv \beta A^\dagger \beta
$$

We have 

$$
\overline{AB} = \beta B^\dagger A^\dagger \beta = \beta B^\dagger\beta\beta A^\dagger\beta =  (\overline{B})(\overline{A})
$$

Also, for a complex number $c$

$$
\overline{cA} = c^* \overline{A}
$$

For a spinor $u$, we define

$$
\overline{u} = u^\dagger\beta
$$


Viewing the Pauli martrices as the Clifford algebra $Cl(3, R)$, define
$$
\sigma^\mu \equiv (I, \vec{\sigma})
$$
$$
\overline{\sigma^\mu} \equiv (I, -\vec{\sigma})
$$

$$
\gamma^\mu \equiv \begin{bmatrix}0 & \sigma^\mu\\ \overline{\sigma^\mu} & 0\end{bmatrix}
$$

Then $\gamma_0 = \beta$ and 

$$
\vec{\gamma} = \begin{bmatrix}0 & \vec{\sigma}\\ -\vec{\sigma} & 0\end{bmatrix}
$$

Note that 

$$
\vec{\gamma}^\dagger = \begin{bmatrix}0 & -\vec{\sigma}\\ \vec{\sigma} & 0\end{bmatrix} = -\vec{\gamma}
$$

and

$$
\{ \gamma^\mu, \gamma^\nu \} = -2 g^{\mu\nu}
$$


$$
\gamma^5 \equiv i\gamma^0\gamma^1\gamma^2\gamma^3 = \begin{bmatrix}-I & 0\\0 & I\end{bmatrix}
$$


A useful property is $\{ \gamma^5, \gamma^\mu \} = 0$

$$
S^{\mu\nu} \equiv \frac{i}{4}[\gamma^\mu, \gamma^\nu]
$$

### Exercise 38.2

Verify 38.15

a) $\overline{\gamma^\mu} = \gamma^\mu$

b) $\overline{S^{\mu\nu}} = S^{\mu\nu}$

c) $\overline{i\gamma^5} = i\gamma^5$

d) $\overline{\gamma^\mu\gamma^5} = \gamma^\mu\gamma^5$

e) $\overline{i\gamma^5 S^{\mu\nu}} = i\gamma^5 S^{\mu\nu}$

#### Solution

a) $\beta \gamma_0 \beta = \beta^3 = \beta^2 \beta = \beta$
whereas,
$$
\beta\vec{\gamma}^\dagger\beta = \begin{bmatrix}0 & I\\I & 0\end{bmatrix} \begin{bmatrix}0 & -\vec{\sigma}\\ \vec{\sigma} & 0\end{bmatrix} \begin{bmatrix}0 & I\\I & 0\end{bmatrix} = \begin{bmatrix}0 & \vec{\sigma}\\ -\vec{\sigma} & 0\end{bmatrix} = \vec{\gamma}
$$

b)
$$
\overline{S^{\mu\nu}} = \overline{\frac{i}{4}[\gamma^\mu, \gamma^\nu]} = 
-\frac{i}{4}(\overline{\gamma^\mu\gamma^\nu} - \overline{\gamma^\nu\gamma^\mu}) = 
-\frac{i}{4} (\gamma^\nu\gamma^\mu - \gamma^\mu\gamma^\nu) = S^{\mu\nu}
$$

c) First, we establish $\gamma^5$ is Hermitian.

$$
(\gamma^5)^\dagger = (i\gamma^0\gamma^1\gamma^2\gamma^3)^\dagger = -i (\gamma^3)^\dagger (\gamma^2)^\dagger  (\gamma^1)^\dagger (\gamma^0)^\dagger
= -i (-\gamma^3)(-\gamma^2)(-\gamma^1)\gamma_0 = i \gamma^3\gamma^2\gamma^1\gamma^0
$$

Now we use anticommuting relations. Since (0123) is an even permutation of (3210), we have 

$$
(\gamma^5)^\dagger = i \gamma^0\gamma^1\gamma^2\gamma^3 = \gamma^5
$$


$$
 \gamma^0  (i\gamma^5)^\dagger\gamma^0 = -i\gamma^0\gamma^5\gamma^0 = (\gamma^0)^2\gamma^1\gamma^2\gamma^3\gamma^0 = \gamma^1\gamma^2\gamma^3\gamma^0 = 
 -\gamma^0\gamma^1\gamma^2\gamma^3 = i\gamma^5
$$

where the anticommuting relation was used in the penultimate step

d) 

$$
\overline{\gamma^\mu\gamma^5} = \overline{\gamma^5} \overline{\gamma^\mu} = i(\overline{i\gamma^5}) \gamma^\mu  =
i  (i\gamma^5)\gamma^\mu = -\gamma^5\gamma^\mu = \gamma^\mu\gamma^5
$$

e) It is easy to see that $\gamma^5[\gamma^\mu, \gamma^\nu] = [\gamma^\mu, \gamma^\nu] \gamma^5$, therefore
$\gamma^5$ commutes with $S^{\mu\nu}$. 

$$
\overline{i\gamma^5 S^{\mu\nu}} = (\overline{S^{\mu\nu}})(\overline{i\gamma^5}) = S^{\mu\nu}i\gamma^5 = i\gamma^5S^{\mu\nu}
$$

### Exercise 38.3

Verify 38.22

a) $\overline{u}_{s'}(p)\beta v_s(-p) = 0$

b) $\overline{v}_{s'}(p)\beta u_s(-p) = 0$


#### Solution


Subtracting 38.20 from 38.19:

$$
\gamma^\mu \slashed{p} - \slashed{p}'\gamma^\mu = -p^\mu - 2iS^{\mu\nu}p_\nu + p'^\mu - 2iS^{\mu\nu}p'_\nu
$$

We sandwich both sides between $\overline{u}_{s'}(p)$ and $v_s(-p)$ 
$$
\overline{u}(p)\left[\gamma^\mu \slashed{p} - \slashed{p}'\gamma^\mu\right]v(-p) = \overline{u}(p)\left[-p^\mu - 2iS^{\mu\nu}p_\nu + p'^\mu - 2iS^{\mu\nu}p'_\nu\right]v(-p)
$$

Using equation 38.1:

$$
\overline{u}(p)\left[-m\gamma^\mu - \slashed{p}'\gamma^\mu\right]v(-p) = \overline{u}(p)\left[-p^\mu - 2iS^{\mu\nu}p_\nu + p'^\mu - 2iS^{\mu\nu}p'_\nu\right]v(-p)
$$

Using equation 38.16:

$$
\slashed{p}\, v(p) = m\, v(p)
$$

$$
\implies (-\slashed{p})\, v(-p) = m\, v(-p)
$$

$$
\implies \slashed{p}\, v(-p) = -m\, v(-p) \tag{38.3.1}
$$


\begin{align}
\overline{u}(p)\left[-m\gamma^\mu + m\gamma^\mu\right]v(-p) &= 
\overline{u}(p)\left[-p^\mu - 2iS^{\mu\nu}p_\nu + p'^\mu - 2iS^{\mu\nu}p'_\nu\right]v(-p) \\
&\implies \overline{u}(p)\left[-p^\mu - 2iS^{\mu\nu}p_\nu + p'^\mu - 2iS^{\mu\nu}p'_\nu\right]v(-p) = 0
\end{align}

$$
[\gamma^{\mu}, \gamma^{\nu}] p_\nu = \gamma^{\mu}\slashed{p} - \gamma^{\nu}\gamma^{\mu}p_\nu
$$

and letting $\mu = 0$, 

$$
\gamma^0\slashed{p} - \gamma^{\nu}\gamma^0 p_\nu = 2\gamma^0\slashed{p} 
$$

$$
\overline{u}(p)\,\gamma^0 \slashed{p}\, v(-p) = 0 \implies
-\overline{u}(p)\,\gamma^0 m\, v(-p) = 0 \implies
\overline{u}_{s'}(p)\,\gamma^0\, v_s(-p) = 0
$$

The other equation is obtained in a similar manner.
