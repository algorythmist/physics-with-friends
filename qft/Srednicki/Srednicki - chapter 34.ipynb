# Representations of $\text{SU}(2) \times \text{SU}(2)$

## The Setup

You have two independent $\text{SU}(2)$ algebras with generators $N_i$ and $N_i^\dagger$ satisfying:

$$[N_i, N_j] = i\varepsilon_{ijk}N_k$$

$$[N_i^\dagger, N_j^\dagger] = i\varepsilon_{ijk}N_k^\dagger$$

$$[N_i, N_j^\dagger] = 0$$

The last equation is the key — the two $\text{SU}(2)$s don't talk to each other.

## Building Representations

Since the algebras are independent, a representation of the product is just a tensor product of representations of each factor:

- Pick a rep of the first $\text{SU}(2)$ with "spin" $n$, dimension $(2n+1)$
- Pick a rep of the second $\text{SU}(2)$ with "spin" $n'$, dimension $(2n'+1)$

The combined representation is labeled $(n, n')$ with:
- **Dimension:** $(2n+1)(2n'+1)$
- **States:** $|n, m; n', m'\rangle$ where $m \in \{-n, \ldots, n\}$ and $m' \in \{-n', \ldots, n'\}$

## Examples

| $(n, n')$ | Dimension | Name |
|:---:|:---:|:---:|
| $(0, 0)$ | $1$ | scalar |
| $\left(\frac{1}{2}, 0\right)$ | $2$ | left-handed spinor |
| $\left(0, \frac{1}{2}\right)$ | $2$ | right-handed spinor |
| $\left(\frac{1}{2}, \frac{1}{2}\right)$ | $4$ | vector |
| $(1, 0)$ | $3$ | self-dual 2-form |
| $(0, 1)$ | $3$ | anti-self-dual 2-form |

## Connection to Spin/Angular Momentum

The physical angular momentum generators are:

$$J_i = N_i + N_i^\dagger$$

Under the rotation subgroup (the diagonal $\text{SU}(2)$), the representation $(n, n')$ decomposes via the usual addition of angular momentum:

$$j = |n - n'|, \, |n - n'| + 1, \, \ldots, \, n + n'$$

So the "spin" of a representation is:

$$j_{\max} = n + n'$$

## Examples of Spin Content

**Vector $\left(\frac{1}{2}, \frac{1}{2}\right)$:** Contains $j = 0, 1$ — a scalar and a 3-vector, i.e., a 4-vector.

**Left-handed spinor $\left(\frac{1}{2}, 0\right)$:** Contains only $j = \frac{1}{2}$.

**$\left(1, \frac{1}{2}\right)$:** Contains $j = \frac{1}{2}, \frac{3}{2}$ — this is a spin-$\frac{3}{2}$ field (like the gravitino).

## Why This Matters for the Lorentz Group

The complexified Lorentz algebra satisfies:

$$\mathfrak{so}(3,1)_{\mathbb{C}} \cong \mathfrak{su}(2) \oplus \mathfrak{su}(2)$$

So finite-dimensional representations of the Lorentz group are classified by pairs $(n, n')$. This is why spinors, vectors, and tensors in relativistic physics are organized this way.
