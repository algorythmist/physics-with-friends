# Lorentz Transformations

Lorentz Transformations is the set of linear transformations that leave invariant the pseudo inner product:

$$
x^2 = g_{mn}x^m x^n = x_n x^n
$$


The Lorentz transformation satisfy

$$
\det L = \pm 1
$$

<b>Proper</b> transformations are the subgroup satisfying $\det L = 1$.
The full group is generated by the proper Lorentz transformations plus Parity and Time Reversal. 

The <b>Orthochornous</b> transformzations are the subgroup satisfying $L^0_0 \geq 1$.
