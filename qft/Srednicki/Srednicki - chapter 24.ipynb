### Problem 24.1 

Let $R = I + \theta + O(\theta^2)$. Show that $\theta$ is antisymmetic

#### Solution

$$
I = R^T R = (I + \theta^T)(I + \theta) = I + \theta^T + \theta + O(\theta^2) \implies \theta^T +\theta = 0
$$


### Problem 24.4
What is the dimension of the space of 2n x 2n matrices that satisfy

$$
S^T\eta S = \eta, \text{where }\eta = \begin{bmatrix}0 & I\\ -I & 0\end{bmatrix}
$$

#### Solution

Since $\eta$ is antisymmetric, so is $S^T\eta S$, so the above equation results in $2n(2n-1)/2 = n(2n-1)$ independent equations.
Therefore the dimension of this matrix space is $(2n)^2 - n(2n-1) = 4n^2 - 2n^2 +n = 2n^2+n = n(2n+1)$
