## General Observations

- If $[A, B] = 0$ then A and B have the same set of eigenvectors. In particular, if A has eigenvalues $a_i$,
then for any analytic function f,
$f(A)$ has the save eigenvectors as A with eigenvalues $f(a_i)$

- A change of basis occurs by applying a unitary operator to an operator A
$A' = U^\dagger AU$. The "rotated" operator $A'$ has the same eigenvalues as A but different eigenvectors.
If $\ket{a_i}$ are the eigenvectors of A, then the eigenvectors of $A'$ are $U^\dagger \ket{a_i}$

- Pick an orthonormal basis $\ket{e_i}$. Then any states can be expressed in terms of these basis:
$$
\ket{a} = \sum_i a_i \ket{e_i}\\
\ket{b} = \sum_i b_i \ket{e_i}
$$

If we compute the coordinates $a_i, b_i$, then we can compute their inner product
$$
\braket{a}{b} = \sum_i \bar{a_i}b_i
$$

In particular, if the state is prepared in $\ket{a}$, 
the probability that the system will be measured in state $\ket{b}$ is $|\braket{a}{b}|^2$

Question

- If $[A,B] = 0$ can we conclude that there is an analytic function f, such that $B = f(A)? 
Analogy with measure theory. If a random variable Y is X-measurable, then there exists a function f 
such that $Y = f(X)$

## Uncertainty

Assuming the system is in state $\psi$

$$
<A> \equiv \bra{\psi}A\ket{\psi}\\
\bar{A} \equiv A - <A>\\
(\Delta A)^2 ≡ <\bar{A}^2>
$$

General uncertainty theorem
$$
\Delta A \Delta B ≥ \frac{1}{2}|<[A,B]>|
$$

### Problem 2.3

Given a general transformation between two bases

$$
\ket{i} = \alpha\ket{u}+\beta\ket{d}
$$

$$
\ket{o} = \gamma\ket{}+\delta\ket{d}
$$

Use the relationships in 2.8 and 2.9 to show that:

a)
$$
|\alpha|^2 = |\beta|^2 = |\gamma|^2 = |\delta|^2 = 1/2
$$

b)
$$
\alpha^*\beta + \alpha\beta^* = \gamma^*\delta + \gamma\delta^* = 0
$$


c) $\alpha^*\beta$  and $\gamma^*\delta$ are pure imaginary

#### Solution

a) Let's start with the relation $\braket{i}{u}\braket{u}{i} = 1/2$ and expand it out

\begin{equation}
1/2 = \braket{i}{u}\braket{u}{i} = 
(\alpha^* \braket{u}{u} + \beta^* \braket{u}{d})(\alpha \braket{u}{u} + \beta \braket{u}{d}) = 
\alpha^* \alpha
\end{equation}

The other equations follow the same way.

b) This time, let's start with $\braket{i}{r}\braket{r}{i} = 1/2$ 

\begin{eqnarray}
1/2 &=& \braket{i}{r}\braket{r}{i}\\
&=& (\alpha^* \braket{u}{r} + \beta^* \braket{d}{r})(\alpha \braket{r}{u} + \beta \braket{r}{d})\\
&=& |\alpha|^2 \braket{u}{r}\braket{r}{u} + |\beta|^2 \braket{d}{r}\braket{r}{d}) +
\alpha^* \beta \braket{u}{r}\braket{r}{d} + \alpha\beta^* \braket{d}{r}\braket{r}{u}\\
&=& 1/2 +1/2(\alpha^* \beta + \alpha\beta^* )
\end{eqnarray}

where I have used equation 2.9 $\braket{x}{r}\braket{r}{x} = 1/2$, where x is anything.
It follows that $\alpha^* \beta + \alpha\beta^*  = 0$.

c) It is in general true that if any complex number satisfies $z+z^* = 0$ then it is pure imaginary.

Just write $z = a+ib$, then  $0 = z+z^* = a+ib+a-ib = 2a \implies a = 0$.

### Problem 3.5

The state is prepared to be in the +1 eigenvalue of $\sigma_m$. The  apparatus measures $\sigma_n$. 
What is the probability that the result is +1?

#### Solution
We can use the expansion in problem 3 to compute $\braket{\sigma_n^+}{\sigma_m^+}$

$$
\cos\frac{\theta_1}{2}\cos\frac{\theta_2}{2}e^{-i\phi_2} + \sin\frac{\theta_1}{2}\sin\frac{\theta_2}{2}e^{-i\phi_1}
$$

Taking the amplitude square, this becomes

$$
\cos^2\frac{\theta_1}{2}\cos^2\frac{\theta_2}{2} + \sin^2\frac{\theta_1}{2}\sin^2\frac{\theta_2}{2}\\
+ \cos\frac{\theta_1}{2}\cos\frac{\theta_2}{2}\sin\frac{\theta_1}{2}\sin\frac{\theta_2}{2}\cos(\phi_1-\phi_2)
$$

To reduce further we must assume that both states have the same orientation in the $\phi$ direction, 
so that $\phi_1 = \phi_2$

$$
\cos^2\frac{\theta_1}{2}\cos^2\frac{\theta_2}{2} + \sin^2\frac{\theta_1}{2}\sin^2\frac{\theta_2}{2}\\
+ \cos\frac{\theta_1}{2}\cos\frac{\theta_2}{2}\sin\frac{\theta_1}{2}\sin\frac{\theta_2}{2} = \\
(cos\frac{\theta_1}{2}\cos\frac{\theta_2}{2} + \sin\frac{\theta_1}{2}\sin\frac{\theta_2}{2})^2 = \\
(cos\frac{\theta_1-\theta_2}{2})^2
$$

### Problem 5.1

Show that all 2x2 Hermitean matrices can be written as 
$$
L = a\sigma_x+b\sigma_y+c\sigma_z+dI
$$
where a,b,c,d are real numbers

#### Solution
The 2x2 Hermitean matrices are precicely the matrices of the form 
$$
L = 
\begin{pmatrix}
r & w\\
w^* & r'
\end{pmatrix}
$$


First note that 

$$
\begin{pmatrix}
r & 0\\
0 & r'
\end{pmatrix} 
= \frac{1}{2}(r+r')
\begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix} 
+\frac{1}{2}(r-r')
\begin{pmatrix}
1 & 0\\
0 & -1
\end{pmatrix}
$$

Now let $w = a-ib$

Then,

$$
\begin{pmatrix}
0 & w\\
w^* & 0
\end{pmatrix}
= 
\begin{pmatrix}
0 & a-ib\\
a+ib & 0
\end{pmatrix}
= 
a
\begin{pmatrix}
0 & 1\\
1 & 0
\end{pmatrix}
+b
\begin{pmatrix}
0 & -i\\
i & 0
\end{pmatrix}
$$

Putting it all together

$$
L = \frac{1}{2}(r+r')I + \frac{1}{2}(r-r')\sigma_z + a\sigma_x+b\sigma_y
$$
