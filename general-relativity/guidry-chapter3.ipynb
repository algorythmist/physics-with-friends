Let 
$$
U^a_b := \partial x'^a/\partial x^b
$$ 
$$
\tilde{U}^b_a := \partial x^b/\partial x'^b
$$

Then 

$$
U^c_a \tilde{U}^b_c = \delta^b_a
$$


The Christoffel symbols are non-tensorial quantities that transform according to

\begin{equation}
\Gamma'^{l}_{ab} = \Gamma^k_{mn}\tilde{U}^m_a\tilde{U}^n_b U^l_k +
U^l_m \frac{\partial^2 x^m}{\partial x'^a \partial x'^b}
\end{equation}

If the manifold has a metric, then they can be defined using the metric as

\begin{equation}
\Gamma^{s}_{lm} = \frac{1}{2}g^{ns} \left( 
g_{mn,l} + g_{ln,m} - g_{ml,v}
\right)
\end{equation}


### Problem 3.1

Show that $T^{ab} = V^aV^b$ transforms as a tensor. Same for lower or mixed indexes.

#### Solution

Let $U^a_b := \partial x'^a/\partial x^b$, 
$\tilde{U}^b_a := \partial x^b/\partial x'^b$

Then $U^c_a \tilde{U}^b_c = \delta^b_a$Let $U^a_b := \partial x'^a/\partial x^b$, 
$\tilde{U}^b_a := \partial x^b/\partial x'^b$

Then $U^c_a \tilde{U}^b_c = \delta^b_a$
\begin{eqnarray}
T'^{ab}e'_a e'_b &=&  V'^a V'^b e'_a e'_b\\
&=& U^a_c V^c U^b_d V^d \tilde{U}^m_a e^m \tilde{U}^n_b e^n\\
&=& U^a_c \tilde{U}^m_a U^b_d \tilde{U}^n_b V^c V^d e^m e^n\\
&=& \delta^m_c \delta^n_d V^c V^d e^m e^n\\
&=& V^c V^d e_c e_d\\
&=& T^{cd}e_c e_d
\end{eqnarray}

### Problem 3.6
Express the flat spacetime metric and volume element in polar coordinates.
Compute the volume of a sphere of radius R.

#### Solution

$$
G = diag(1, -1, -r^2, -r^2\sin^2\theta)
$$

$$
dV =\sqrt{|g|}d^4 x = r^2\sin\theta d^4 x
$$

For a sphere

$$
V = \int r^2\sin\theta dr d\theta d\phi = 
\int dr r^2 \int d\theta \sin\theta \int d\phi =
 \frac{4}{3}\pi R^3
$$

### Problem 3.13

Accept the following *Quotient Theorem*:

If a set of quantities produces a tensor when contracted with a tensor,
then that set of quantities is a tensor.

Use the quotient theorem to argue that the metric tensor is a tensor.

#### Solution


We know that the scalar product $A\cdot B = g_{\mu\nu}A^{\mu}B^{\nu}$ is invariant under coordinate transformations. We also know that the tensor product of two vectors 
$A^{\mu}B^{\nu}$ is tensor. From the quotient theorem, we can conclude $g_{\mu\nu}$ is also a tensor.

### Problem 3.17

Prove equation (3.67)

#### Solution

Let $x'^m = x^m +X^m(x)\delta u$.

\begin{eqnarray}
T'^{mn}(x') &=& \frac{\partial x'^m}{\partial x^a}\frac{\partial x'^n}{\partial x^b}T^{ab}(x)\\
&=& \left( \frac{\partial x^m}{\partial x^a} + X^m_{,a} \delta u \right)
\left( \frac{\partial x^n}{\partial x^b} + X^n_{,b} \delta u \right)
T^{ab}\\
&=&  \left( \delta^m_a\delta^n_b + \delta^m_a X^n_{,b}\delta u +
\delta^n_b X^m_{,a}\delta u + O(\delta u^2) \right) T^{ab}\\
&=&  T^{mn} + \left( T^{mb}X^a_{,b} + T^{an}X^m_{,a} \right)
\end{eqnarray}

On the other hand, the Taylor expansion of $T^{mn}(x')$ around $x$ is

\begin{equation}
T^{mn}(x') = T^{mn}(x) + X^a T^{mn}_{,a}(x)
\end{equation}

By the definition of the Lie derivative 

$$
\mathcal{L}_X T^{mn} = \lim_{\delta u \rightarrow 0} \frac{T^{mn}(x') - T'^{mn}(x')}{\delta u}
$$

and substituting the two expressions we got above, we get

$$
\mathcal{L}_X T^{mn} = X^a T^{mn}_{,a} - T^{mb}X^n_{,b} - T^{an} X^m_{,a}
$$


### Problem 3.18
Derive the Lie derivative of a vector.

#### Solution

As in the previous problem, the relevant transformation is $x'^m = x^m +X^m(x)\delta u$, from which

$$
V'^m(x') = \frac{\partial x'^m}{\partial x^b} V^b(x) = 
\left( 
\frac{\partial x^m}{\partial x^b} +\frac{\partial X^m}{\partial x^b}\delta u
\right) V^b(x) = V^m(x) + X^m_{,b}V^b(x)\delta u
$$

We also have by Taylor expansion

$$
V^m(x') = V^m(x)+X^a\partial_a V^m \delta u = V^m(x)+X^a V^m_{,a}  \delta u 
$$

Putting these expressions in the definition of the Lie derivative we get

$$
\mathcal{L}_X V^m = \lim_{\delta u \rightarrow 0} \frac{V^m(x') - V'^m(x')}{\delta u} = X^aV^m_{,a} - V^b X^m_{,b}
$$


### Problem 3.20

Show that the Lie bracket of two vector fields $[A,B]$ yield a vector field equal to $\mathcal{L}_A B$.

#### Solution

\begin{eqnarray*}
[A,B]f &=& [A^{\mu}\partial_\mu, B^{\nu}\partial_\nu]f\\ 
&=& A^{\mu}\partial_\mu(B^{\nu}\partial_\nu f) - 
B^{\mu}\partial_\mu(A^{\nu}\partial_\nu f)\\
&=& A^{\mu}\partial_{\mu} B^{\nu} \partial_{\nu}f - 
B^{\mu}\partial_\mu A^{\mu} \partial_{\nu}f\\
&=& ( A^{\mu}\partial_{\mu} B^{\nu}  - B^{\mu}\partial_\mu A^{\mu})\partial_{\nu}f
\end{eqnarray*}


### Problem 3.23

Show that the Lie derivative for a vector field remains the same if partial derivatives are replaced with covariant derivatives.

#### Solution

The Lie derivate for a vector field is given by

$$
\mathcal{L}_X V^m = X^aV^m_{,a} - V^b X^m_{,b}
$$

The covariant derivative of a vector is given by

$$
V^m_{;a} = V^m_{,a} + \Gamma^m_{ca}V^c
$$

Similarily

$$
X^m_{;b} = X^m_{,b} + \Gamma^m_{cb}X^c
$$

So

$$
X^aV^m_{;a} - V^b X^m_{;b} =  X^a V^m_{,a} + \Gamma^m_{ca}X^aV^c - V^b X^m_{,b} - \Gamma^m_{cb}X^c V^b 
= \mathcal{L}_X V^m + \Gamma^m_{ca}X^aV^c - \Gamma^m_{cb}X^c V^b 
$$

Interchanging the lower indices on $\Gamma$ causes the last two terms to cancel out.
