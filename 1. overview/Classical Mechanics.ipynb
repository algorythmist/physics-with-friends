# Lagrangian formulation

## Single Particle

Minimizing the action

$$
S = \int L(q, \dot{q}) \,dt
$$

yields the Euler Lagrange equations:

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot{q}} - \frac{\partial L}{\partial q} = 0
$$


Define $p := \frac{\partial L}{\partial \dot{q}}$, and define the Hamiltonian by

$$
H = p\dot{q} - L
$$

To get

$$
\frac{\partial H}{\partial p } = \dot{q} ; \frac{\partial H}{\partial q } = -\dot{p}
$$

## Many Particles

Minimizing the action

$$
S = \int \sum_i L(q_i, \dot{q_i}) \,dt \tag{1}
$$

yields the Euler Lagrange equations:

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot{q_i}} - \frac{\partial L}{\partial q_i} = 0 \tag{2}
$$


Define $p := \frac{\partial L}{\partial \dot{q}}$, and define the Hamiltonian by

$$
H = \sum_i p_i\dot{q_i} - L
$$



## Fields

### 1-D Field

$q_i(t) \rightarrow \phi(x,t)$

$\sum_i \rightarrow \int dx$


$$
S = \int L(q_i, \dot{q_i}) \,dt \rightarrow \int dt dx \mathcal{L}[\phi(x,t), \partial_t \phi(x,t), \partial_x \phi(x,t)]
$$


The index $i$ turns into the continuous index $x$. $q_i(t)$ becomes $\phi(t, x)$. 
Setting $x^0 = t$, ww can simply write $\phi(x)$ in 4D spacetime.

Notation:
$$
\phi_{,\mu} \equiv \partial_\mu \phi \equiv \frac{\partial \phi}{\partial  x^\mu}
$$

Then (1) becomes

$$
S = \int dt \int d^3 x \mathcal{L}(\phi, \phi_{,\mu})
$$


The Euler Lagrange equations (2) become

$$
\partial_{\mu} \frac{\partial \mathcal{L}}{\partial \phi_{,\mu}} - \frac{\partial \mathcal{L}}{\partial\phi} = 0
$$

Energy-Momentum density and Hamiltonian

$$
\Pi^\mu := \frac{\partial \mathcal{L}}{\partial \phi_{,\mu}}
$$

$$
\pi(x,t) = \Pi^0(x,t) = \partial_0 \phi(x,t) = \dot{\phi}
$$


## Noether's Theorem

### Continuous Symmetry

Consider a transformation $\varphi_a(x) \rightarrow \varphi_a(x) + \delta\varphi_a(x)$

Then

$$
\delta\mathcal{L}(x) = \frac{\partial\mathcal{L}}{\partial\varphi_a(x)} \delta\varphi_a(x) + \frac{\partial\mathcal{L}}{\partial(\partial_\mu\varphi_a(x))} \partial_\mu\delta\varphi_a(x)
$$

Replace the first term using the EL equations

$$
\delta\mathcal{L}(x) = \partial_\mu\left(\frac{\partial\mathcal{L}(x)}{\partial(\partial_\mu\varphi_a(x))} \delta\varphi_a(x)\right) + \frac{\delta S}{\delta\varphi_a(x)}\delta\varphi_a(x) 
$$

Define

$$
j^\mu = \frac{\partial\mathcal{L}(x)}{\partial(\partial_\mu\varphi_a(x))} \delta\varphi_a(x)  \text{   (summation over a)}
$$

$$
Q = \int d^3 x j^0(x)
$$

Then 

$$
\partial_\mu j^\mu =  \delta\mathcal{L}(x) -  \frac{\delta S}{\delta\varphi_a(x)}\delta\varphi_a(x)
$$

If the transformation leaves the action unchanged, then $\partial_\mu j^\mu =  0$
