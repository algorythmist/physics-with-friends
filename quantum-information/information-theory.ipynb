**In Markov inequality, why does the r.v. have to be strictly positive?**

Look at the chain of inequalities for general $X \in \mathbb{R}$, which could be negative.

$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x p(x)\, dx 
= \int_{-\infty}^{-a} x p(x)\, dx 
+ \int_{-a}^{a} x p(x)\, dx 
+ \int_{a}^{\infty} x p(x)\, dx
$$

Label:
- (i) $\int_{-\infty}^{-a} x p(x)\, dx$
- (ii) $\int_{-a}^{a} x p(x)\, dx$
- (iii) $\int_{a}^{\infty} x p(x)\, dx$

→ Replacing $x \to a$ in the integrand of (iii) still gives the inequality:

$$
\mathbb{E}[X] 
\geq \int_{-\infty}^{-a} x p(x)\, dx 
+ \int_{-a}^{a} x p(x)\, dx 
+ a \int_{a}^{\infty} p(x)\, dx
$$

Removing term (ii) gives:

$$
\mathbb{E}[X] 
\geq \int_{-\infty}^{-a} x p(x)\, dx 
+ a \int_{a}^{\infty} p(x)\, dx
$$

→ This is the step that gets us into trouble if $p(x) > 0$ for $x < 0$.

We want:

$$
\mathbb{E}[X] \geq a \int_{a}^{\infty} p(x)\, dx \tag{★}
$$

But if $X$ can be $< 0$, then term (i) is negative and we can no longer claim (★).

**Example:**

If $e = a + b$, we can only claim $e \geq b$ if $a \geq 0$.

Or in numbers:

$$
5 \geq 4 + 0.9 \Rightarrow 5 \geq 4 \Rightarrow 5 > 0.9
$$

But:

$$
5 \geq -20 + 17 \quad \text{but} \quad 5 \not\geq 17 \Rightarrow \text{negative term ruins it}
$$


```python
def gcd(i, j):
  while j != 0:
    i , j = j, i % j
  return i

def choose(n, k):
  a, b = 1, 1
  while k > 0:
    a *= n
    b *= k
    d = gcd(a, b)
    if d != 1:
      a //= d
      b //= d
    n -= 1
    k -= 1
  if b != 1:
    raise ValueError
  return a
```

```python
import matplotlib.pyplot as plt

def show(n, p):
  x = list(range(n + 1))
  y1 = [choose(n, i) for i in x]
  y2 = [p ** i * (1 - p) ** (n - i) for i in x]
  y3 = [c * z for c, z in zip(y1, y2)]
  plt.plot(x, y1)
  plt.xlabel('Heads')
  plt.ylabel('Count')
  plt.title('Coin Flip Count')
  plt.show()
  plt.plot(x, y3)
  plt.xlabel('Heads')
  plt.ylabel('Probability')
  plt.title('Binomial Distribution (p = %.1f)' % p)
  plt.show()
```

```python
show(10, 0.6)
```

```python
show(100, 0.6)
```

```python
show(1000, 0.6)
```
